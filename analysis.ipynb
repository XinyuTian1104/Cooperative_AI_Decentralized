```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
loss_lists = pickle.load(open("loss_lists.pkl", "rb"))
reward_lists = pickle.load(open("reward_lists.pkl", "rb"))
information = pickle.load(open("information.pkl", "rb"))
```

```python
# plot avg reward
rewards = [np.array(reward_lists[x][-1]).mean() for x in range(len(reward_lists))]
plt.plot(rewards)
plt.title("Average Reward")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()

```

```python
# plot avg loss
losses = [np.array(loss_lists[x][-1]).mean() for x in range(len(loss_lists))]
plt.plot(losses)
plt.title("Average Loss")
plt.xlabel("Episode")
plt.ylabel("Loss")
plt.show()
```
